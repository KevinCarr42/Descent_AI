```python
class Temp:
    def __init__(self, combined_map, label_map):
        self.combined_map = combined_map
        self.label_map = label_map

    def show(self):
        SCALE = 0.5
        x, y = self.combined_map.shape
        
        cmap = ListedColormap(COLOUR_DF.colour.to_list())
        norm = BoundaryNorm(
            COLOUR_DF.id.to_list(), 
            cmap.N, 
            clip=True
        )
        
        plt.figure(figsize=(SCALE*y, SCALE*x))
        plt.pcolor(
            self.combined_map,
            cmap=cmap,
            norm=norm,
            edgecolors='k',
            linewidth=3,
        )
        
        # ticks and labels
        ax = plt.gca()
        ax.invert_yaxis()
        ax.set_aspect('equal')
        ax.xaxis.tick_top()        
        plt.xticks([i+0.5 for i in range(y)], [i+1 for i in range(y)])
        plt.yticks([i+0.5 for i in range(x)], [i+1 for i in range(x)])
        
        # labels
        # TODO
        
        plt.show()

Temp(combined_map.fillna(0), label_map).show()
```
